# Dataset Description

The dataset is selected from the Kaggle 

https://www.kaggle.com/datasets/jiteshkumarsahoo/wikipedia-country-statistics-2023/data

This dataset provides a comprehensive snapshot of global country statistics for the year 2023. It was scraped from various Wikipedia pages using BeautifulSoup, consolidating key indicators and metrics for 142 countries. The dataset covers diverse aspects such as land area, water area, Human Development Index (HDI), GDP forecasts, internet usage, and population changes.


### Key Columns and Metrics:

 1. Country: The name of the country.
 2. Total in km2: Total area of the country.
 3. Land in km2: Land area excluding water bodies.
 4. Water in km2: Area covered by water bodies.
 5. Water %: Percentage of the total area covered by water.
 6. HDI: Human Development Index, a measure of a country's overall achievement in its social and economic dimensions.
%
 7. HDI Growth: Percentage growth in HDI.
 8. IMF Forecast GDP(Nominal): International Monetary Fund's forecast for Gross Domestic Product in nominal terms.
 9. World Bank Forecast GDP(Nominal): World Bank's forecast for Gross Domestic Product in nominal terms.
 10.UN Forecast GDP(Nominal): United Nations' forecast for Gross Domestic Product in nominal terms.
 11. IMF Forecast GDP(PPP): IMF's forecast for Gross Domestic Product in purchasing power parity terms.
 12. World Bank Forecast GDP(PPP): World Bank's forecast for Gross Domestic Product in purchasing power parity terms.
 13. CIA Forecast GDP(PPP): Central Intelligence Agency's forecast for Gross Domestic Product in purchasing power parity terms.
 14. Internet Users: Number of internet users in the country.
 15. UN Continental Region: Continental region classification by the United Nations.
 16. UN Statistical Subregion: Statistical subregion classification by the United Nations.
 17. Population 2022: Population of the country in the year 2022.
 18. Population 2023: Population of the country in the year 2023.
 19. Population %Change: Percentage change in population from 2022 to 2023.

```python
#importing necessary libraries for visualization
import matplotlib as plt

import matplotlib.pyplot as plt

import seaborn as sns

# imports for offline figures 
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# imports to view figures in jupyter notebooks
#from plotly.offline import init_notebook_mode
#init_notebook_mode(connected = True)
# you may need this instead
import plotly.io as pio
pio.renderers.default = "notebook_connected"


#import for publishing online
from chart_studio.plotly import plot
```

### 1. Which countries have the highest population in 2023?

```python
# Identify countries with the highest population in 2023
top_countries = data.nlargest(10, 'Population 2023')['Country']

top_countries

```

```python
# Plot Choropleth for countries with the highest population in 2023
fig_choropleth = px.choropleth(locations=top_countries, locationmode='country names',
                               color=top_countries, title='Choropleth: Countries with Highest Population in 2023',
                              )

# Show the Choropleth map
fig_choropleth.show()
```

 The Choropleth map represents the spatial distribution of the top 10 countries with the highest population in 2023. India has the largest population in the world in the year of 2023.

### 2. Which countries are the top contributors to high water distribution?

```python
# Create a bar graph
fig_bar = px.bar(data, x='Country',
                 y='Water %', 
                 title='Water Distribution by Country')

# Sort the DataFrame by 'Water %' in descending order and select the top 10
highest = data.sort_values(by='Water %', ascending=False).head(10)

# Create a new bar graph showing only the top 10 countries
fig_bar = px.bar(highest, x='Country',
                 y='Water %',
                 title='Top 10 Countries with Highest Water Distribution',
                color_discrete_sequence=["rgb(65,187,197)"])

# Show the chart for the top 10 countries
fig_bar.show()
```

The bar chart represents the top 10 countries with highest water distribution.Bahamas which is a Country in the Caribbean having the highest water distribution.

### 3. What is the total water Volume by Continental Regions

```python

# Calculate 'body_water' column
data['body_water'] = data['Total in km2'] * (data['Water %'] / 100)

# Create a bar chart with Plotly Express
fig = px.bar(
    data,
    x='UN Continental Region',
    y='body_water',
    title='Water Volume in Km2 by Region',
    labels={'UN Continental Region': 'UN Continental Region', 'body_water': 'Water Volume (Km2)'},
    color='UN Continental Region', 
    color_continuous_scale='Magma',
)

# Update layout
fig.update_layout(
    xaxis_title='UN Continental Region',
    yaxis_title='Water Volume (Km2)',
    showlegend=True,
   
)

# Show the plot
fig.show()
```

chart representing the water volume in square kilometers by UN Continental Region, using the 'body_water' column calculated from the 'Total in km2' and 'Water %' columns in the dataset.The UN continental region Americas having the highest water volume when compared to other continental regions.

### 4. What is the the relationship between the Human Development Index (HDI) and the International Monetary Fund's forecast for Gross Domestic Product (GDP) in nominal terms for different countries

```python
# Create a scatter plot for HDI vs. GDP relationship
fig_scatter = px.scatter(data, x='HDI', y='IMF Forecast GDP(Nominal)',
                         color='Country', size='Population 2023',
                         title='HDI vs. GDP Relationship for Different Countries',
                         labels={'IMF Forecast GDP': 'GDP Forecast'},
                         template='plotly_dark')

# Show the scatter plot
fig_scatter.show()
```



This scatter plot visualizes the relationship between the Human Development Index (HDI) and the International Monetary Fund's forecast for Gross Domestic Product (GDP) in nominal terms for different countries. Each point represents a country, with the x-axis showing HDI, the y-axis showing GDP forecast, and the size of the points indicating the population in 2023. The color distinguishes between different countries. Analyzing this plot can provide insights into the correlation between a country's human development and its economic forecast.

### 7. Which country has the highest GDP forecast according to the IMF, World Bank, and UN?

```python
# Sort the DataFrame by IMF Forecast GDP(PPP) in descending order
data_sorted = data.sort_values(by="IMF Forecast GDP(PPP)", ascending=False)

# Select the top 3 countries
top_countries = data_sorted.head(3)
top_countries
```

```python

gdp_forecast_cols = ["IMF Forecast GDP(Nominal)", "World Bank Forecast GDP(Nominal)", "UN Forecast GDP(Nominal)"]

fig = make_subplots(rows=1, cols=3, subplot_titles=gdp_forecast_cols)

colors=["rgb(65,187,197)","rgb(32,80,46)","rgb(141,235,113)"]

# Iterate over each GDP forecast column
for i, col in enumerate(gdp_forecast_cols, start=1):
    # Add a trace for each country in a subplot
    for j, country in enumerate(top_countries["Country"]):
        fig.add_trace(go.Bar(x=[country],
                             y=[top_countries.loc[top_countries["Country"] == country, col].values[0]],
                             name=country,
                             marker_color=colors[j]),
                      row=1, col=i)

    
fig.update_layout(height=400, width=1200, 
                  title_text="Top 3 Countries GDP Forecasts Comparison",
                  showlegend=False)

# Show the plot
fig.show()
```

Selecting the top 3 countries by IMF Forecast GDP(PPP)and ploting them according to IMF Forecast GDP(Nominal),World Bank Forecast GDP(Nominal),UN Forecast GDP(Nominal)as subplot.In the terms of IMF Forecast GDP(Nominal) Germany having the highest GDP Forcasts and in terms of World Bank Forecast and UN Forecast GDP(Nominal) Japan having the highest GDP. 


### 8. What is the comprehensive relationship between a country's Human Development Index (HDI), various aspects of its Gross Domestic Product (GDP) including forecasts (Nominal and PPP), Population dynamics?

```python
# GDP Growth Trends
# Correlation heatmap for %HDI Growth, Population %Change and GDP forecasts
gdp_growth_data = data[['%HDI Growth', 'Population %Change', 'IMF Forecast GDP(Nominal)',
                             'World Bank Forecast GDP(Nominal)', 'CIA Forecast GDP(PPP)']]
corr = gdp_growth_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for HDI Growth, Population Change and GDP Forecasts')
plt.show()
```

From the correlation matrix its clear that some of these features are highly correlated to each other at the same time some of them are highly negatively correlated to eachother.The parameters like World Bank Forecast GDP and IMF Forecast GDP are highly correlated to each other at the same time World Bank Forecast FDP and IMF Forecast GDP are negatively correlated with the HDI growth percentage.
